```python
from sklearn import datasets
from sklearn import svm
from PIL import Image, ImageTk
#from cx_Freeze import setup, Executable
import pandas as pd
import numpy as np
import joblib
```

```python
import tkinter
```

```python
def exibir_imagem(nome_flor, janela):
    nome_arquivo = f"{nome_flor.lower()}.png"

    try:
        imagem = Image.open(nome_arquivo)
        imagem = imagem.resize((150, 150), resample=Image.Resampling.LANCZOS)
        imagem = ImageTk.PhotoImage(imagem)
        imagem_label.configure(image=imagem)
        imagem_label.image = imagem
    except Exception as e:
        print(f"Erro ao abrir a imagem: {e}")

def classificar():
    sl = float(sepala_l.get())
    sa = float(sepala_a.get())
    pl = float(petala_l.get())
    pa = float(petala_l.get())
    dados = np.array([[sa, sl, pa, pl]])
    botanico = joblib.load('modelo.pkl')
    classificacao = botanico.predict(dados)
    if classificacao == 0:
        resultado['text'] = 'Iris-Setosa'
        exibir_imagem('iris_setosa', janela)
    elif classificacao == 1:
        resultado['text'] = 'Iris-Versicolor'
        exibir_imagem('iris_versicolor', janela)
    else:
        resultado['text'] = 'Iris-Versicolor'
        exibir_imagem('iris_virginica', janela)

janela = tkinter.Tk()
janela.geometry('800x600')
janela.title('Classificador de Flores')
janela.configure(bg='#ebebeb')  # Cor de fundo cinza claro

rotulo_opcoes = tkinter.Label(janela, text="Digite as opções de Comprimento e Largura", font=('Helvetica', 14), bg='#ebebeb')
rotulo_opcoes.pack(pady=10)

# Frame para Sépala
frame_sepala = tkinter.Frame(janela, bg='#c5c5c5')
frame_sepala.pack(pady=10)

label_sepala_a = tkinter.Label(frame_sepala, text="Comprimento Sépala:", bg='#484848', fg='white')
label_sepala_a.pack(side=tkinter.LEFT)

sepala_a = tkinter.Entry(frame_sepala)
sepala_a.pack(side=tkinter.LEFT)

label_sepala_l = tkinter.Label(frame_sepala, text="Largura Sépala:", bg='#484848', fg='white')
label_sepala_l.pack(side=tkinter.LEFT)

sepala_l = tkinter.Entry(frame_sepala)
sepala_l.pack(side=tkinter.LEFT)

# Frame para Pétala
frame_petala = tkinter.Frame(janela, bg='#c5c5c5')
frame_petala.pack(pady=10)

label_petala_a = tkinter.Label(frame_petala, text="Comprimento Pétala:", bg='#484848', fg='white')
label_petala_a.pack(side=tkinter.LEFT)

petala_a = tkinter.Entry(frame_petala)
petala_a.pack(side=tkinter.LEFT)

label_petala_l = tkinter.Label(frame_petala, text="Largura Pétala:", bg='#484848', fg='white')
label_petala_l.pack(side=tkinter.LEFT)

petala_l = tkinter.Entry(frame_petala)
petala_l.pack(side=tkinter.LEFT)

# Botão para classificar
botao = tkinter.Button(janela, text='Que flor é essa?', command=classificar, bg='#484848', fg='white')
botao.pack()

# Rótulo para exibir o resultado
resultado = tkinter.Label(janela, text="")
resultado.pack()

# Rótulo para exibir a imagem
imagem_label = tkinter.Label(janela)
imagem_label.pack(pady=10)


janela.mainloop()
```
