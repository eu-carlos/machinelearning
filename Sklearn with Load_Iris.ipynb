```python
# importing libs

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing dataset from sklearn

from sklearn.datasets import load_iris
data = load_iris()
```

```python
# dataset characteristics

print(data['DESCR'])
```

```python
data

# dataset information
```

```python
# maps numeric target values to class names.
def defineNames(target):
    class_names = {0: "setosa", 1: "versicolor", 2: "virginica"}
    return class_names.get(target, "Unknown class name")
```

```python
# Getting the dataset
iris = pd.DataFrame(data.data, columns=["sepal length", "sepal width", "petal length", "petal width"])
```

```python
iris["target"] = data.target
iris["target_name"] = iris["target"].apply(defineNames)
```

```python
# display the DataFrame
display(iris)
```

```python
# display descriptive statistics
iris.describe()
```

```python
import seaborn as sns
sns.set(rc={"figure.figsize":(15,12)})
sns.set_theme(style="white")
```

```python
ax = sns.boxplot(data=iris)
```

```python
# checking null values
iris.info()
```

```python
# select only numeric columns for correlation analysis, creating and displaying correlation matrix
numeric_columns = iris.select_dtypes(include=[float, int]).columns
correlation_matrix = iris[numeric_columns].corr()
print(correlation_matrix)
```

```python
ax = sns.pairplot(iris)
```

```python
ax = sns.pairplot(iris, hue="target_name")
```

```python
# scatter plot: petal width vs. petal length, colored by target species
sns.scatterplot(x="petal width", y="petal length", data=iris, hue="target_name", palette=["green", "orange", "blue"])

# labels and title
plt.xlabel("Width")
plt.ylabel("Length")
plt.title("Petal Characteristics by Species")

# display plot
plt.show()
```

```python
# import train_test from sklearn
from sklearn.model_selection import train_test_split

# spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=1)
```

```python
# convert the training to a DataFrame
train = pd.DataFrame(X_train)

# rename the columns
train.columns = ["sepal length", "sepal width", "petal length", "petal width"]

# add the target column to the training DataFrame
train["target"] = y_train
```

```python
display(train)
```

```python
# scatter plot to visualize petal width vs. petal length in the training set, colored by target species
sns.scatterplot(x="petal width", y="petal length", data=train, hue="target", palette=["green", "orange", "blue"])

# adding labels and title
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Scatter Plot: Petal Width vs. Petal Length in Training Set")

# display the legend for target species
plt.legend(title="Target Species", labels=["Setosa", "Versicolor", "Virginica"])

# show the plot
plt.show()
```

```python
# filter training data with specific petal widths
train[(train["target"] == 2) & ((train["petal width"] == 1.7) | (train["petal width"] == 1.5))]
```

```python
# scatter plot to visualize petal width vs. petal length in the training set, colored by target species
sns.scatterplot(x="petal width", y="petal length", data=train, hue="target", palette=["green", "orange", "blue"])

# adding labels and title
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Scatter Plot: Petal Width vs. Petal Length in Training Set")

# draw a line to emphasize a trend for the green points
sns.lineplot(x=[0, 1.4], y=[4.8, 0], color="green", label="Green Trend")

# draw a line to emphasize a trend for the blue points
sns.lineplot(x=[0, 4.9], y=[7.7, 0], color="blue", label="Blue Trend")

# display the legend
plt.legend()

# show the plot
plt.show()
```

```python
# Convert the test to a DataFrame
test = pd.DataFrame(X_test)

# rename the columns
test.columns = ["sepal length", "sepal width", "petal length", "petal width"]

# add the target column to the testing DataFrame
test["target"] = y_test
```

```python
display(test)
```

```python
# scatter plot to visualize petal width vs. petal length in the testing set, colored by target species
sns.scatterplot(x="petal width", y="petal length", data=test, hue="target", palette=["green", "orange", "blue"])

# adding labels and title
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Scatter Plot: Petal Width vs. Petal Length in Testing Set")

# draw a line to emphasize a trend for the green points
X_green = [0, 1.4]
y_green = [4.8, 0]
sns.lineplot(x=X_green, y=y_green, color="green", label="Green Trend")

# draw a line to emphasize a trend for the blue points
X_blue = [0, 4.9]
y_blue = [7.7, 0]
sns.lineplot(x=X_blue, y=y_blue, color="blue", label="Blue Trend")

# display the legend
plt.legend()

# show the plot
plt.show()
```

```python
# import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
# train the RandomForestClassifier on the training data
model.fit(X_train,y_train)
```

```python
# make predictions on the test data
prediction = model.predict(X_test)
```

```python
prediction
```

```python
# add predicted labels and a check column to the testing DataFrame
test["prediction"] = prediction
test["check"] = test["target"] == test["prediction"]

# display the testing DataFrame
display(test)
```

```python
# Show incorrect predictions
test[test.check == False]
```

```python
from sklearn.metrics import accuracy_score

# calculate accuracy
accuracy = accuracy_score(test["target"], test["prediction"])

# print the accuracy
print(f"Model Accuracy: {accuracy:.2%}")
```
